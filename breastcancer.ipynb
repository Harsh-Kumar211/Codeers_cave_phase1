### Encoding converting objects types into int or float

```python
df = pd.get_dummies(df,columns=['Race','Marital Status'], dtype=int, drop_first=True)
```

```python
df.rename(columns={"T Stage ": "T Stage"}, inplace=True)
df["Grade"].replace({" anaplastic; Grade IV": "4"}, inplace=True)
df["Grade"] = df["Grade"].astype(int)
df["T Stage"].replace({"T1":0, "T2": 1, "T3":2, "T4": 3}, inplace=True)
df["N Stage"].replace({"N1":0, "N2": 1, "N3":2}, inplace=True)
df["6th Stage"].replace({"IIA":0, "IIB": 1, "IIIA":2, "IIIB": 3,"IIIC":4}, inplace=True)
df["differentiate"].replace({"Moderately differentiated": 2,
                            "Poorly differentiated": 1,
                            "Well differentiated": 3,
                            "Undifferentiated": 0}, inplace=True)
df["A Stage"].replace({"Regional":0, "Distant": 1}, inplace=True)
df["Estrogen Status"].replace({"Positive":0, "Negative": 1}, inplace=True)
df["Progesterone Status"].replace({"Positive":0, "Negative": 1}, inplace=True)
df["Status"].replace({"Alive":0, "Dead": 1}, inplace=True)
```

### Data Cleaning

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
lowage,uppage=remove_outlier(df['Survival Months'])
df['Survival Months']=np.where(df['Survival Months']>uppage,uppage,df['Survival Months'])
df['Survival Months']=np.where(df['Survival Months']<lowage,lowage,df['Survival Months'])
```

```python
sns.boxplot(x='Survival Months',data=df)
```

```python
sns.boxplot(x='Tumor Size',data=df)
```

```python
lowage,uppage=remove_outlier(df['Tumor Size'])
df['Tumor Size']=np.where(df['Tumor Size']>uppage,uppage,df['Tumor Size'])
df['Tumor Size']=np.where(df['Tumor Size']<lowage,lowage,df['Tumor Size'])
```

```python
sns.boxplot(x='Tumor Size',data=df)
```

```python
X=df.drop(columns = 'Status')
y=df['Status']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
predictions=lr.predict(X_train)
cm=metrics.confusion_matrix(y_train,predictions, labels=[1,0])
df_cm=pd.DataFrame(cm,index=[i for i in ["1","0"]],
                   columns=[i for i in ["Predict1","Predict 0"]])
sns.heatmap(df_cm,annot=True,fmt='g')
plt.title('heat map of train data')
plt.show()
```

```python
predictions=lr.predict(X_test)
cm=metrics.confusion_matrix(y_test,predictions, labels=[1,0])
df_cm=pd.DataFrame(cm,index=[i for i in ["1","0"]],
                   columns=[i for i in ["Predict1","Predict 0"]])
sns.heatmap(df_cm,annot=True,fmt='g')
plt.title('heat map of test data')
plt.show()
```
